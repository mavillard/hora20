```python
import os

from bs4 import BeautifulSoup
```

```python
DIR_DATA = 'data/html/'
```

```python
os.listdir(DIR_DATA)[:10]
```

```python
%%time

files_topics_dict = {}

for filename in os.listdir(DIR_DATA)[:3]:
    with open(DIR_DATA + filename) as f:
        topics_dict = {}
        date = filename # Extraer fecha del filename
        soup = BeautifulSoup(f, 'html.parser')
        topics = [a.string for a in soup.select('ul.vbs-topics-list li a')]
        for topic in topics:
            if topic != 'ALL TOPICS':
                topics_dict[topic] = []
                lis = soup.select('ul[tid="' + topic + '"] li.key')
                for li in lis:
                    kw_freq = (li.find('a').string.strip(), li.find('span').string.strip('()'))
                    topics_dict[topic].append(kw_freq)
        files_topics_dict[filename] = topics_dict
```
